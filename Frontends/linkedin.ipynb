
## express service



### entry

linkedin service?


```python


const path = require('path')
const express = require('express')
const cors = require('cors')

async function startServer() {
  var app = express()
  app.use(cors({
    credentials: true,
    origin: '*' 
  }))
  app.use(express.json({ limit: '20mb' }))
  app.get('/styles.css', function (req, res) {
    res.setHeader('content-type', 'text/css');
    res.send(importer.interpret('linkedin stylesheet').code)
  })
  app.get('/profile.css', function (req, res) {
    res.setHeader('content-type', 'text/css');
    res.send(importer.interpret('linkedin profile stylesheet').code)
  })
  app.get('/homepage.css', function (req, res) {
    res.setHeader('content-type', 'text/css');
    res.send(importer.interpret('linkedin homepagestylesheet').code)
  })
  app.get('/in/:profile', function (req, res) {
    return res.send(importer.interpret('linkedin index').code
      .replaceAll('{STYLESHEETS}', '<link rel="stylesheet" href="/profile.css?t={TIMESTAMP}">')
      .replaceAll('{MAINCONTENT}', importer.interpret('linkedin profile').code))
  })
  app.get('/', function (req, res) {
    return res.send(importer.interpret('linkedin index').code
      .replaceAll('{STYLESHEETS}', '<link rel="stylesheet" href="/homepage.css?t={TIMESTAMP}">')
      .replaceAll('{MAINCONTENT}', importer.interpret('linkedin homepage').code))
  })

  let port = process.env.GITHUB_PORT || 9191
  app.listen(port, () => console.log('Listening on https://localhost:' + port))
}

module.exports = startServer


```
