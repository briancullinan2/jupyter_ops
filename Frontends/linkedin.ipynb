```python
<!-- Profile Card -->
<form method="post" action="/profile">
  <div class="container">
    <div class="profile">
      <div id="drop-banner" class="profile-banner" style="background-image: url({BANNER});"></div>
      <div id="drop-area" class="profile-info">
        <img id="preview" src="{PICTURE}" alt="Profile Photo">
        <input type="file" id="fileElem" accept="image/*" style="display:none;">
        <button onclick="document.getElementById('fileElem').click()">Choose File</button>
        <h2><input name="name" placeholder="First and last name" value="{NAME}" /></h2>
        <p><input name="title" placeholder="Job title" value="{TITLE}" /></p>
        <p><input name="location" placeholder="Location" value="{LOCATION}"></p>
      </div>
    </div>
    <div class="sidebar">
      <h3>Contact info</h3>
      <p><a href="{URL}">{URL}</a></p>
      <p><input name="phone" placeholder="Phone not provided" value="{PHONE}" /></p>
      <p><input name="email" placeholder="Email address" value="{EMAIL}" /></p>
      <button type="submit">Save</button>
    </div>
  </div>

  <!-- About Section -->
  <div class="section">
    <h3>About</h3>
    <p><textarea name="summary" placeholder="Summary about you">{SUMMARY}</textarea></p>
  </div>

  <!-- Experience Section -->
  <div class="section">
    <h3>Experience</h3>
    {EXPERIENCE}
    <button type="submit" name="submit" value="add">Add</button>
  </div>

  <!-- Education Section -->
  <div class="section">
    <h3>Education</h3>
    {EDUCATION}
    <button type="submit" name="submit" value="add">Add</button>
  </div>

  <!-- Skills Section -->
  <div class="section">
    <h3>Skills</h3>
    <div class="skills">
      {SKILLS}
    </div>
    <button type="submit" name="submit" value="add">Add</button>
  </div>

</form>

<script>
  const dropArea = document.getElementById('drop-area');
  const dropBanner = document.getElementById('drop-banner');
  const preview = document.getElementById('preview');

  function uploadFile(file, banner) {
    const reader = new FileReader();
    reader.onload = function () {
      const base64 = reader.result;
      if(banner) {
        dropBanner.style.backgroundImage = 'url(' + base64 + ')';
      } else {
        preview.src = base64;
      }

      fetch('/upload', {
        method: 'POST',
        credentials: 'include',
        headers: { 'Content-Type': 'application/json' },
        body: JSON.stringify(banner ? { banner: base64 } : { image: base64 })
      })
        .then(res => res.text())
        //.then(msg => alert(msg))
        .catch(err => console.error('Upload failed:', err));
    };
    reader.readAsDataURL(file);
  }

  dropArea.addEventListener('dragover', e => {
    e.preventDefault();
    dropArea.classList.add('hover');
  });

  dropArea.addEventListener('dragleave', () => {
    dropArea.classList.remove('hover');
  });

  dropArea.addEventListener('drop', e => {
    e.preventDefault();
    dropArea.classList.remove('hover');
    const file = e.dataTransfer.files[0];
    if (file && file.type.startsWith('image/')) {
      uploadFile(file);
    }
  });

  dropBanner.addEventListener('dragover', e => {
    e.preventDefault();
    dropBanner.classList.add('hover');
  });

  dropBanner.addEventListener('dragleave', () => {
    dropBanner.classList.remove('hover');
  });

  dropBanner.addEventListener('drop', e => {
    e.preventDefault();
    dropBanner.classList.remove('hover');
    const file = e.dataTransfer.files[0];
    if (file && file.type.startsWith('image/')) {
      uploadFile(file, true);
    }
  });

  document.getElementById('fileElem').addEventListener('change', e => {
    const file = e.target.files[0];
    if (file && file.type.startsWith('image/')) {
      uploadFile(file);
    }
  });
</script>
```

```python
const {ERRORS, SESSIONS} = importer.import('serve linkedin login')
const fs = require('fs')
const path = require('path')
const crypto = require('crypto')

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')

const PROFILES = {}

function getProfile(session_id) {

  let session
  let profile
  let homepage = 'linkedin login page'
  if(typeof SESSIONS[session_id] != 'undefined') {
    homepage = 'linkedin homepage'
    session = SESSIONS[session_id]
    if(PROFILES[session.filename]) {
      profile = PROFILES[session.filename]
    } else {
      profile = JSON.parse(fs.readFileSync(path.join(LINKEDIN_PATH, session.filename)))
      PROFILES[session.filename] = profile
    }

    if(!profile) {
      return {}
    } else {
      profile = profile[0]
    }
  }

  return {session, profile}
}

module.exports = {
  getProfile,
  PROFILES,
}
```

### homepage

serve linkedin homepage?

```python
const {ERRORS, SESSIONS} = importer.import('serve linkedin login')
const {getProfile, PROFILES} = importer.import('linkedin load profile')
const fs = require('fs')
const path = require('path')
const crypto = require('crypto')

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')

function linkedinHomepage (req, res) {
  if(!req.cookies.session_id) {
    res.cookie('session_id', (req.cookies.session_id = crypto.randomBytes(16).toString('hex')))
  }

  let {session, profile} = getProfile(req.cookies.session_id)
  let homepage = 'linkedin login page'
  if(typeof SESSIONS[req.cookies.session_id] != 'undefined') {
    homepage = 'linkedin homepage'
  }

  if(req.query.code) {
    ERRORS[req.cookies.session_id] = 'Enter a new password'
  }

  let printed = ERRORS[req.cookies.session_id]
  ERRORS[req.cookies.session_id] = ''

  return res.send(importer.interpret('linkedin index').code
    .replaceAll('{STYLESHEETS}', '<link rel="stylesheet" href="/linkedin-home.css?t={TIMESTAMP}">')
    .replaceAll('{TIMESTAMP}', Date.now())
    .replaceAll('{PAGE}', typeof SESSIONS[req.cookies.session_id] != 'undefined' ? 'home' : 'login')
    .replaceAll('{MYPROFILE}', session ? ('<a href="/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', '') + '">My Profile</a>') : '')
    .replaceAll('{MAINCONTENT}', importer.interpret(homepage).code
      .replaceAll('{ERRORS}', printed || '')
      .replaceAll('{NAME}', profile ? profile.name || '': '')
      .replaceAll('{TITLE}', profile ? profile.title || '' : '')
      .replaceAll('{CODE}', req.query.code || '')
      .replaceAll('{PROFILE}', session ? ('/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', '')) : '')
    ))
}

module.exports = linkedinHomepage

```

```python

const fs = require('fs')
const path = require('path')
const { safeurl } = importer.import('domain cache tools')
const crypto = require('crypto')
var nodemailer = require('nodemailer');

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')
const PASSWD_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn/passwd.json')
const SALT_PATH = path.join(HOMEPATH, '.credentials/linkedin-salt.json')

const SESSIONS = fs.existsSync(PASSWD_PATH) ? JSON.parse(fs.readFileSync(PASSWD_PATH)) : {}
const ERRORS = {}
const SALT = JSON.parse(fs.readFileSync(SALT_PATH))
const ITERATIONS = 1 << 10

function generateSHA512(input) {
  const hash = crypto.createHash('sha512')
  hash.update(input)
  return hash.digest('hex')
}

function sendReset(session_id) {
  
  if(SESSIONS[session_id]) {
    SESSIONS[session_id].verify = crypto.randomBytes(16).toString('hex')
    fs.writeFileSync(PASSWD_PATH, JSON.stringify(SESSIONS, null, 4))

    var transporter = nodemailer.createTransport({
      host: 'smtp.gmail.com', // hostname
      secure: false, // use SSL
      port: 587,
      //port: 465,
      //port: 1587, // port for secure SMTP
      auth: {
        user: process.env.ADMIN_USER,
        pass: process.env.ADMIN_PASS,
      },
      tls: {
          rejectUnauthorized: false
      }
    })
    var mailOptions = {
        from:'megamindbrian@gmail.com',
        to: SESSIONS[session_id].email,
        subject: 'Reset Linked Out password',
        text:'Click here to reset: https://sosh.pryor.games/?code=' + SESSIONS[session_id].verify
    }
    transporter.sendMail(mailOptions, (error, info) => {
      if (error) {
          console.log('Error sending email:', error);
      } else {
          console.log('Email sent:', info.response);
      }
    })
  }

}

function linkedinLogin(req, res, next) {
  if (!req.cookies.session_id) {
    throw new Error('Cookies not enabled')
  }

  // TODO: find already listed filename/email address association
  let sessions = Object.keys(SESSIONS)
  let session
  for (let i = 0; i < sessions.length; i++) {
    if (SESSIONS[sessions[i]].email == req.body.username) {
      session = SESSIONS[sessions[i]]
      if(req.body.submit == 'reset') {
        sendReset(sessions[i])
      }    
      break;
    }
  }

  if(req.body.submit == 'reset') {
    ERRORS[req.cookies.session_id] = 'Email sent'
    return res.redirect('/')
  }

  if (!session && req.body.submit != 'register') {
    ERRORS[req.cookies.session_id] = 'No user account'
    return res.redirect('/')
  } else if (!session) {
    session = {}
  }

  // TODO: more validation
  // TODO: check password

  let count = ITERATIONS
  let encodedPassword
  let salt
  if(req.body.submit == 'register') {
    salt = crypto.randomBytes(16).toString('hex')
  } else {
    salt = session.salt
  }
  encodedPassword = generateSHA512(salt + req.body.password)
  do {
    encodedPassword = generateSHA512(encodedPassword + SALT + req.body.password)
    count--
  } while (count > 0)


  if(req.body.submit == 'register') {

    if(!req.body.username.match(/^\w+([-+.']\w+)*@\w+([-.]\w+)*\.\w+([-.]\w+)*$/)) {
      ERRORS[req.cookies.session_id] = 'No user account'
      return res.redirect('/')
    }

    session.salt = salt
    session.email = req.body.username
    session.password = encodedPassword
    session.filename = 'linkedin-' + safeurl(req.body.first + ' ' + req.body.last) + '.json'
    session.verify = crypto.randomBytes(16).toString('hex')
    let existCount = 1
    let safetyCount = 10000
    while(fs.existsSync(path.join(LINKEDIN_PATH, session.filename)) && safetyCount > 0) {
      session.filename = session.filename.replace(/[0-9]*.json$/gi, '') + existCount + '.json'
      existCount++
      safetyCount--
    }
    if(safetyCount == 0) {
      ERRORS[req.cookies.session_id] = 'No user account'
      return res.redirect('/')
    }
    
    // first time file is being created
    fs.writeFileSync(path.join(LINKEDIN_PATH, session.filename), JSON.stringify([{
      name: req.body.first + ' ' + req.body.last,
      email: req.body.username
    }], null, 4))

    // send verification email
    sendVerification(session.verify, session.email)

    SESSIONS[req.cookies.session_id] = session
  } else if (encodedPassword != session.password && session.verify != req.body.code) {
    // return non-specific error for security
    ERRORS[req.cookies.session_id] = 'No user account'
    return res.redirect('/')
  } else {
    // TODO: set this if login successful
    if(session.verify == req.body.code) {
      session.password = encodedPassword
      session.verify = true // because it goes through email
    }
    SESSIONS[req.cookies.session_id] = session
  }

  fs.writeFileSync(PASSWD_PATH, JSON.stringify(SESSIONS, null, 4))

  res.redirect('/?t=' + Date.now())
}

module.exports = {
  linkedinLogin,
  SESSIONS,
  ERRORS,
}
```

```python

const {ERRORS, SESSIONS} = importer.import('serve linkedin login')
const fs = require('fs')
const path = require('path')
const crypto = require('crypto')
const {getProfile, PROFILES} = importer.import('linkedin load profile')


const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')

function linkedinJobs (req, res) {
  if(!req.cookies.session_id) {
    res.cookie('session_id', (req.cookies.session_id = crypto.randomBytes(16).toString('hex')))
  }

  let {session, profile} = getProfile(req.cookies.session_id)
  if(!session || !profile) {
    ERRORS[req.cookies.session_id] = 'Login to see jobs'
    return res.redirect('/')
  }

  // TODO: list all profile files
  let profiles = fs.readdirSync(LINKEDIN_PATH)
  let jobFiles = []
  for(let i = 0; i < profiles.length; i++) {
    if(req.path.includes('/jobs')) {
      if(profiles[i].startsWith('jobs-') && profiles[i].endsWith('.json')) {
        jobFiles.push(profiles[i].replace('jobs-', '').replace('.json', ''))
        if(jobFiles.length == 16) {
          break
        }
      }
    } else {
      // create mode matches personal jobs listing
      if(profiles[i] == session.filename.replace('linkedin-', 'jobs-')) {
        jobFiles.push(profiles[i].replace('jobs-', '').replace('.json', ''))
        break;
      }
    }
  }

  let jobTemplate
  if(req.path.includes('/create')) {
    jobTemplate = importer.interpret('linkedin jobs edit item').code
  } else {
    jobTemplate = importer.interpret('linkedin jobs list item').code
  }
  let jobCount = 0
  let jobs = jobFiles.map(profileFile => {
    let jobs = JSON.parse(fs.readFileSync(path.join(LINKEDIN_PATH, 'jobs-' + profileFile + '.json')))
    return jobs.map(job => jobTemplate
      .replaceAll('{TITLE}', job.title || '')
      .replaceAll('{DESCRIPTION}', job.description || '')
      .replaceAll('{QUALIFICATIONS}', job.qualifications || '')
      .replaceAll('{DATES}', job.dates || '')
      .replaceAll('{INDEX}', jobCount++)
    ).join('\n')
  })
  if(req.path.includes('/create')) {
    jobs.push(jobTemplate
      .replaceAll('{TITLE}', '')
      .replaceAll('{DESCRIPTION}', '')
      .replaceAll('{QUALIFICATIONS}', '')
      .replaceAll('{DATES}', '')
      .replaceAll('{INDEX}', jobs.length)
    )
  }

  let mainTemplate = 'linkedin edit jobs template'
  if(req.path.includes('/jobs')) {
    mainTemplate = 'linkedin jobs template'
  }

  return res.send(importer.interpret('linkedin index').code
    .replaceAll('{STYLESHEETS}', '<link rel="stylesheet" href="/linkedin-network.css?t={TIMESTAMP}">')
    .replaceAll('{PAGE}', req.path.includes('/jobs') ? 'jobs' : 'create')
    .replaceAll('{MYPROFILE}', SESSIONS[req.cookies.session_id] ? ('<a href="/in/' + path.basename(SESSIONS[req.cookies.session_id].filename).replace('linkedin-', '').replace('.json', '') + '">My Profile</a>') : '')
    .replaceAll('{MAINCONTENT}', importer.interpret(mainTemplate).code
      .replaceAll('{JOBS}', jobs.join('\n'))
    ))
}

module.exports = linkedinJobs

```
