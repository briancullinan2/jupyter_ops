```python
<!-- Profile Card -->
<form method="post" action="/profile">
  <div class="container">
    <div class="profile">
      <div id="drop-banner" class="profile-banner" style="background-image: url({BANNER});"></div>
      <div id="drop-area" class="profile-info">
        <img id="preview" src="{PICTURE}" alt="Profile Photo">
        <input type="file" id="fileElem" accept="image/*" style="display:none;">
        <button onclick="document.getElementById('fileElem').click()">Choose File</button>
        <h2><input name="name" placeholder="First and last name" value="{NAME}" /></h2>
        <p><input name="title" placeholder="Job title" value="{TITLE}" /></p>
        <p><input name="location" placeholder="Location" value="{LOCATION}"></p>
      </div>
    </div>
    <div class="sidebar">
      <h3>Contact info</h3>
      <p><a href="{URL}">{URL}</a></p>
      <p><input name="phone" placeholder="Phone not provided" value="{PHONE}" /></p>
      <p><input name="email" placeholder="Email address" value="{EMAIL}" /></p>
      <button type="submit">Save</button>
    </div>
  </div>

  <!-- About Section -->
  <div class="section">
    <h3>About</h3>
    <p><textarea name="summary" placeholder="Summary about you">{SUMMARY}</textarea></p>
  </div>

  <!-- Experience Section -->
  <div class="section">
    <h3>Experience</h3>
    {EXPERIENCE}
    <button type="submit" name="submit" value="add">Add</button>
  </div>

  <!-- Education Section -->
  <div class="section">
    <h3>Education</h3>
    {EDUCATION}
    <button type="submit" name="submit" value="add">Add</button>
  </div>

  <!-- Skills Section -->
  <div class="section">
    <h3>Skills</h3>
    <div class="skills">
      {SKILLS}
    </div>
    <button type="submit" name="submit" value="add">Add</button>
  </div>

</form>

<script>
  const dropArea = document.getElementById('drop-area');
  const dropBanner = document.getElementById('drop-banner');
  const preview = document.getElementById('preview');

  function uploadFile(file, banner) {
    const reader = new FileReader();
    reader.onload = function () {
      const base64 = reader.result;
      if(banner) {
        dropBanner.style.backgroundImage = 'url(' + base64 + ')';
      } else {
        preview.src = base64;
      }

      fetch('/upload', {
        method: 'POST',
        credentials: 'include',
        headers: { 'Content-Type': 'application/json' },
        body: JSON.stringify(banner ? { banner: base64 } : { image: base64 })
      })
        .then(res => res.text())
        //.then(msg => alert(msg))
        .catch(err => console.error('Upload failed:', err));
    };
    reader.readAsDataURL(file);
  }

  dropArea.addEventListener('dragover', e => {
    e.preventDefault();
    dropArea.classList.add('hover');
  });

  dropArea.addEventListener('dragleave', () => {
    dropArea.classList.remove('hover');
  });

  dropArea.addEventListener('drop', e => {
    e.preventDefault();
    dropArea.classList.remove('hover');
    const file = e.dataTransfer.files[0];
    if (file && file.type.startsWith('image/')) {
      uploadFile(file);
    }
  });

  dropBanner.addEventListener('dragover', e => {
    e.preventDefault();
    dropBanner.classList.add('hover');
  });

  dropBanner.addEventListener('dragleave', () => {
    dropBanner.classList.remove('hover');
  });

  dropBanner.addEventListener('drop', e => {
    e.preventDefault();
    dropBanner.classList.remove('hover');
    const file = e.dataTransfer.files[0];
    if (file && file.type.startsWith('image/')) {
      uploadFile(file, true);
    }
  });

  document.getElementById('fileElem').addEventListener('change', e => {
    const file = e.target.files[0];
    if (file && file.type.startsWith('image/')) {
      uploadFile(file);
    }
  });
</script>
```

### entry

linkedin service?


```python
const linkedinHomepage = importer.import('serve linkedin homepage')
const linkedinProfile = importer.import('serve linkedin profile')
const {linkedinLogin} = importer.import('serve linkedin login')
const linkedinSave = importer.import('save linkedin profile')
const linkedinPic = importer.import('save linkedin pic')
const linkedinBlob = importer.import('linkedin profile blob')

const express = require('express')
const cors = require('cors')
const cookieParser = require('cookie-parser');

async function startServer() {
  var app = express()
  app.use(cors({
    credentials: true,
    origin: '*' 
  }))
  app.use(cookieParser());
  app.use(express.json({ limit: '20mb' }))
  app.use(express.urlencoded({ extended: true }));
  app.get('/styles.css', function (req, res) {
    res.setHeader('content-type', 'text/css');
    res.send(importer.interpret('linkedin stylesheet').code)
  })
  app.get('/profile.css', function (req, res) {
    res.setHeader('content-type', 'text/css');
    res.send(importer.interpret('linkedin profile stylesheet').code)
  })
  app.get('/homepage.css', function (req, res) {
    res.setHeader('content-type', 'text/css');
    res.send(importer.interpret('linkedin homepagestylesheet').code)
  })
  app.get('/in/:profile/profile.jpg', linkedinBlob)
  app.get('/in/:profile/banner.jpg', linkedinBlob)
  app.post('/upload', linkedinPic)
  app.post('/profile', linkedinSave)
  app.post('/login', linkedinLogin)
  app.get('/edit/:profile', linkedinProfile)
  app.get('/in/:profile', linkedinProfile)
  app.get('/', linkedinHomepage)

  let port = process.env.GITHUB_PORT || 9191
  app.listen(port, () => console.log('Listening on https://localhost:' + port))
}

module.exports = startServer


```

### homepage

serve linkedin homepage?


```python
const {ERRORS, SESSIONS} = importer.import('serve linkedin login')
const fs = require('fs')
const path = require('path')
const crypto = require('crypto')

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')

function linkedinHomepage (req, res) {
  if(!req.cookies.session_id) {
    res.cookie('session_id', (req.cookies.session_id = crypto.randomBytes(16).toString('hex')))
  }

  let session
  let profile
  let homepage = 'linkedin login page'
  if(typeof SESSIONS[req.cookies.session_id] != 'undefined') {
    homepage = 'linkedin homepage'
    session = SESSIONS[req.cookies.session_id]
    profile = JSON.parse(fs.readFileSync(path.join(LINKEDIN_PATH, session.filename)))

    if(!profile) {
      return next()
    } else {
      profile = profile[0]
    }
  }

  let printed = ERRORS[req.cookies.session_id]
  ERRORS[req.cookies.session_id] = ''

  return res.send(importer.interpret('linkedin index').code
    .replaceAll('{STYLESHEETS}', '<link rel="stylesheet" href="/homepage.css?t={TIMESTAMP}">')
    .replaceAll('{PAGE}', typeof SESSIONS[req.cookies.session_id] != 'undefined' ? 'home' : 'login')
    .replaceAll('{MAINCONTENT}', importer.interpret(homepage).code
      .replaceAll('{TIMESTAMP}', Date.now())
      .replaceAll('{ERRORS}', printed || '')
      .replaceAll('{NAME}', profile ? profile.name || '': '')
      .replaceAll('{TITLE}', profile ? profile.title || '' : '')
      .replaceAll('{PROFILE}', session ? ('/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', '')) : '')
    ))
}

module.exports = linkedinHomepage

```


### profile

serve linkedin profile?


```python
const fs = require('fs')
const path = require('path')
const {SESSIONS} = importer.import('serve linkedin login')

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')


function linkedinProfile(req, res, next) {
  // TODO: lookup profile in filesystem
  let profiles = fs.readdirSync(LINKEDIN_PATH)
  let profile
  let profileFile
  for(let i = 0; i < profiles.length; i++) {
    if(profiles[i].startsWith('linkedin-') && profiles[i].endsWith('.json') 
      && profiles[i].includes(req.params['profile'])) {
      profileFile = profiles[i]
      break
    }
  }

  if(!profileFile) {
    return next()
  } else {
    profile = JSON.parse(fs.readFileSync(path.join(LINKEDIN_PATH, profileFile)))
    profile = profile[0] || {}
  }

  console.log(profile)
  let template = 'linkedin profile'
  let experienceTemplate = 'linkedin experience view row'
  let educationTemplate = 'linkedin education view row'
  if(req.path.includes('/edit/')) {
    template = 'linkedin edit'
    experienceTemplate = 'linkedin experience edit row'
    educationTemplate = 'linkedin education edit row'
  }

  let experience = ''
  let experienceCount = 0
  let education = ''
  let educationCount = 0
  if(profile.experience) {
    let hasExperience = false
    for(let i = 0; i < profile.experience.length; i++) {
      if(profile.experience[i].length == 1) {
        hasExperience = false
      }

      if(profile.experience[i][0] == 'Experience') {
        hasExperience = true
        continue
      }

      if(!hasExperience) {
        continue
      }

      let newTemplate = importer.interpret(experienceTemplate).code
        .replaceAll('{INDEX}', experienceCount)

      if(profile.experience[i][1] == 'Company Name') {
        newTemplate = newTemplate.replaceAll('{ROLE}', profile.experience[i][0])
                                 .replaceAll('{COMPANY}', profile.experience[i][2])
      }

      if(profile.experience[i][3] == 'Dates Employed') {
        newTemplate = newTemplate.replaceAll('{DATES}', profile.experience[i][4])
      }


      if(profile.experience[i][7] == 'Location') {
        newTemplate = newTemplate.replaceAll('{LOCATION}', profile.experience[i][8])
      }

      if(profile.experience[i][9]) {
        newTemplate = newTemplate.replaceAll('{SUMMARY}', profile.experience[i][9])
      }

      experience += newTemplate
      experienceCount++
    }

    let hasEducation = false
    for(let i = 0; i < profile.experience.length; i++) {
      if(profile.experience[i].length == 1) {
        hasEducation = false
      }

      if(profile.experience[i][0] == 'Education') {
        hasEducation = true
        continue
      }

      if(!hasEducation) {
        continue
      }


      let newTemplate = importer.interpret(educationTemplate).code
        .replaceAll('{INDEX}', educationCount)

      if(profile.experience[i][1] == 'Degree Name') {
        newTemplate = newTemplate.replaceAll('{INSTITUTION}', profile.experience[i][0])
                                 .replaceAll('{CERTIFICATION}', profile.experience[i][2])
      }

      if(profile.experience[i][5].includes('Dates')) {
        newTemplate = newTemplate.replaceAll('{DATES}', profile.experience[i][6])
      }

      education += newTemplate
      educationCount++
    }
  }

  // insert an extra empty on at the bottom for adding
  if(req.path.includes('/edit/') && experienceCount < 99) {
    let newTemplate = importer.interpret(experienceTemplate).code
      .replaceAll('{INDEX}', experienceCount)
      .replaceAll('{ROLE}', '')
      .replaceAll('{COMPANY}', '')
      .replaceAll('{DATES}', '')
      .replaceAll('{LOCATION}', '')
      .replaceAll('{SUMMARY}', '')
    experience += newTemplate
  }
  if(req.path.includes('/edit/') && educationCount < 99) {
    newTemplate = importer.interpret(educationTemplate).code
      .replaceAll('{INDEX}', educationCount)
      .replaceAll('{INSTITUTION}', '')
      .replaceAll('{CERTIFICATION}', '')
      .replaceAll('{DATES}', '')
    education += newTemplate
  }

  let skills = ''
  if(profile.skills) {
    skills = profile.skills.map(skill => skill[0])
      .filter((a, i, arr) => arr.indexOf(a) == i)
      .map(skill => '<span class="skill-tag">' + skill + '</span>')
  }
  if(req.path.includes('/edit/')) {
    skills += '<input name="skill' + skills.length + '" placeholder="Add skill" />'
  }

  return res.send(importer.interpret('linkedin index').code
    .replaceAll('{STYLESHEETS}', '<link rel="stylesheet" href="/profile.css?t={TIMESTAMP}">')
    .replaceAll('{PAGE}', req.path.includes('/edit/') ? 'edit' : 'view')
    .replaceAll('{MAINCONTENT}', importer.interpret(template).code
      .replaceAll('{SKILLS}', skills)
      .replaceAll('{SUMMARY}', profile.summary || '')
      .replaceAll('{LOCATION}', profile.location || '')
      .replaceAll('{NAME}', profile.name || '')
      .replaceAll('{TITLE}', profile.title || '')
      .replaceAll('{URL}', profile.url || '/in/' + path.basename(profileFile).replace('linkedin-', '').replace('.json', ''))
      .replaceAll('{PHONE}', profile.phone || '')
      .replaceAll('{EMAIL}', profile.email)
      .replaceAll('{PICTURE}', profile.picture || '')
      .replaceAll('{BANNER}', profile.banner || '')
      .replaceAll('{SUMMARY}', profile.summary || '')
      .replaceAll('{EXPERIENCE}', experience)
      .replaceAll('{EDUCATION}', education)
      .replaceAll('{EDITABLE}', SESSIONS[req.cookies.session_id] && profileFile == SESSIONS[req.cookies.session_id].filename ? ('<div class="button"><a href="/edit/' + path.basename(profileFile).replace('linkedin-', '').replace('.json', '') + '">Edit</a></div>') : '')
    ))
}

module.exports = linkedinProfile
```

```python

const fs = require('fs')
const path = require('path')
const { safeurl } = importer.import('domain cache tools')
const crypto = require('crypto')

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')
const PASSWD_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn/passwd.json')
const SALT_PATH = path.join(HOMEPATH, '.credentials/linkedin-salt.json')

const SESSIONS = fs.existsSync(PASSWD_PATH) ? JSON.parse(fs.readFileSync(PASSWD_PATH)) : {}
const ERRORS = {}
const SALT = JSON.parse(fs.readFileSync(SALT_PATH))
const ITERATIONS = 1 << 10

function generateSHA512(input) {
  const hash = crypto.createHash('sha512')
  hash.update(input)
  return hash.digest('hex')
}

function linkedinLogin(req, res, next) {
  if (!req.cookies.session_id) {
    throw new Error('Cookies not enabled')
  }

  // TODO: find already listed filename/email address association
  let sessions = Object.keys(SESSIONS)
  let session
  for (let i = 0; i < sessions.length; i++) {
    if (SESSIONS[sessions[i]].email == req.body.username) {
      session = SESSIONS[sessions[i]]
      break;
    }
  }

  if(req.body.submit == 'reset') {
    
    ERRORS[req.cookies.session_id] = 'Email sent'
    return res.redirect('/')
  }

  if (!session && req.body.submit != 'register') {
    ERRORS[req.cookies.session_id] = 'No user account'
    return res.redirect('/')
  } else if (!session) {
    session = {}
  }

  // TODO: more validation
  // TODO: check password

  let count = ITERATIONS
  let encodedPassword
  let salt
  if(req.body.submit == 'register') {
    salt = crypto.randomBytes(16).toString('hex')
  } else {
    salt = session.salt
  }
  encodedPassword = generateSHA512(salt + req.body.password)
  do {
    encodedPassword = generateSHA512(encodedPassword + salt + req.body.password)
    count--
  } while (count > 0)


  if(req.body.submit == 'register') {
    session.salt = salt
    session.email = req.body.username
    session.password = encodedPassword
    session.filename = 'linkedin-' + safeurl(req.body.first + ' ' + req.body.last) + '.json'
    let existCount = 1
    let safetyCount = 10000
    while(fs.existsSync(path.join(LINKEDIN_PATH, session.filename)) && safetyCount > 0) {
      session.filename = session.filename.replace(/[0-9]+$/gi, '') + existCount
      existCount++
      safetyCount--
    }
    if(safetyCount == 0) {
      ERRORS[req.cookies.session_id] = 'No user account'
      return res.redirect('/')
    }
    fs.writeFileSync(path.join(LINKEDIN_PATH, session.filename), JSON.stringify([{
      name: req.body.first + ' ' + req.body.last,
      email: req.body.username
    }]))
    SESSIONS[req.cookies.session_id] = session
  } else if (encodedPassword != session.password) {
    // return non-specific error for security
    ERRORS[req.cookies.session_id] = 'No user account'
    return res.redirect('/')
  } else {
    // TODO: set this if login successful
    SESSIONS[req.cookies.session_id] = session
  }

  fs.writeFileSync(PASSWD_PATH, JSON.stringify(SESSIONS))

  res.redirect('/?t=' + Date.now())
}

module.exports = {
  linkedinLogin,
  SESSIONS,
  ERRORS,
}
```

```python
const fs = require('fs')
const path = require('path')
const {SESSIONS, ERRORS} = importer.import('serve linkedin login')
const moment = require('moment');

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')

function calculateDateDifference(startDate, endDate, unit = 'days') {
  //const format = "MMM YYYY"
  const start = moment(startDate);
  const end = moment(endDate);
  return end.diff(start, unit);
}

function linkedinSave(req, res) {

  let session
  let profile
  if(typeof SESSIONS[req.cookies.session_id] != 'undefined') {
    session = SESSIONS[req.cookies.session_id]
    profile = JSON.parse(fs.readFileSync(path.join(LINKEDIN_PATH, session.filename)))

    if(!profile) {
      ERRORS[req.cookies.session_id] = 'No user account'
      return res.redirect('/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', ''))
    } else {
      profile = profile[0] || {}
    }
  } else {
    ERRORS[req.cookies.session_id] = 'No user account'
    return res.redirect('/')
  }

  // TODO: skip old information, but keep the same format
  let oldExperience = []
  if(profile.experience) {
    let hasExperience = false
    let hasEducation = false
    for(let i = 0; i < profile.experience.length; i++) {
      if(profile.experience[i].length == 1) {
        hasExperience = false
        hasEducation = false
      }
      if(profile.experience[i][0] == 'Experience') {
        hasExperience = true
      }
      if(profile.experience[i][0] == 'Education') {
        hasEducation = true
      }

      if(hasExperience || hasEducation) {
        continue
      }

      oldExperience.push(profile.experience[i])
    }
  }
  
  let newExperience = [
    ['Experience']
  ]

  // TODO: extract form data
  let experienceCount = 0
  while(req.body['role' + experienceCount] && experienceCount < 99) {

    const differenceInMonths = calculateDateDifference(
      req.body['dates' + experienceCount].split(' - ')[0], 
      req.body['dates' + experienceCount].split(' - ')[1], 
      'months');

    newExperience.push([
      req.body['role' + experienceCount],
      'Company Name',
      req.body['company' + experienceCount],
      'Dates Employed',
      req.body['dates' + experienceCount],
      'Employment Duration',
      Math.floor(differenceInMonths / 12) + ' yrs, ' + (differenceInMonths % 12) + ' mos',
      'Location',
      req.body['location' + experienceCount],
      req.body['summary' + experienceCount]
    ])
    experienceCount++
  }

  newExperience.push(['Education'])

  let educationCount = 0
  while(req.body['institution' + educationCount] && educationCount < 99) {
    newExperience.push([
      req.body['institution' + educationCount],
      "Degree Name",
      req.body['certification' + educationCount].split(' in ')[0],
      "Field Of Study",
      req.body['certification' + educationCount].split(' in ')[1],
      "Dates attended or expected graduation",
      req.body['attended' + educationCount]
    ])
    educationCount++
  }

  let oldSkills = profile.skills || []
  let flatSkills = oldSkills.map(skill => skill[0])
  // TODO: leave old skills in tact and add non existent skills

  let newSkills = []
  // TODO: extract form data in sequence
  let skillCount = 0
  while(req.body['skill' + skillCount] && skillCount < 99) {
    // if(!flatSkills.includes())
    newSkills.push([req.body['skill' + skillCount]])
    skillCount++
  }

  // TODO: save all the new information in order
  fs.writeFileSync(path.join(LINKEDIN_PATH, session.filename), JSON.stringify([Object.assign(profile, {
    skills: oldSkills.concat(newSkills),
    experience: oldExperience.concat(newExperience),
    name: req.body.name,
    title: req.body.title,
    location: req.body.location,
    email: req.body.email,
    phone: req.body.phone,
    summary: req.body.summary,
    
  })]))

  if(req.body.submit == 'add') {
    res.redirect('/edit/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', ''))
  } else {
    res.redirect('/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', ''))
  }
}

module.exports = linkedinSave
```

save linkedin pic?


```python
const fs = require('fs')
const path = require('path')
const {SESSIONS, ERRORS} = importer.import('serve linkedin login')
const {spawnSync} = require('child_process')
const mime = require('mime-types')

const HOMEPATH = process.env.HOME || process.env.HOMEPATH || process.env.USERPROFILE
const LINKEDIN_PATH = path.join(HOMEPATH, 'Documents/Collections/conversations/LinkedIn')

function linkedinPic(req, res) {

  let session
  let profile
  if(typeof SESSIONS[req.cookies.session_id] != 'undefined') {
    session = SESSIONS[req.cookies.session_id]
    profile = JSON.parse(fs.readFileSync(path.join(LINKEDIN_PATH, session.filename)))

    if(!profile) {
      ERRORS[req.cookies.session_id] = 'No user account'
      return res.redirect('/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', ''))
    } else {
      profile = profile[0] || {}
    }
  } else {
    ERRORS[req.cookies.session_id] = 'No user account'
    return res.redirect('/')
  }

  let uploadImage = req.body.image
  if(req.body.banner) {
    uploadImage = req.body.banner
  }
  let payload = uploadImage.split(/[:;,]/gi)
  let imageData = payload[3]
  let extension = mime.extension(payload[1]);
  let profileImage = path.join(LINKEDIN_PATH, session.filename).replace('.json', '.' + extension)
  if(req.body.banner) {
    profileImage = path.join(LINKEDIN_PATH, session.filename).replace('.json', '-banner.' + extension)
  }

  fs.writeFileSync(profileImage, Buffer.from(imageData, 'base64'))

  if(path.extname(profileImage) != '.jpeg') {
    spawnSync('convert', [profileImage, profileImage.replace(path.extname(profileImage), '.jpeg')], {
      cwd: LINKEDIN_PATH,
      stdio: [0, 1, 2]
    })
  }

  if(req.body.banner) {
    fs.writeFileSync(path.join(LINKEDIN_PATH, session.filename), JSON.stringify([Object.assign(profile, {
      banner: '/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', '')
        + '/banner.jpg'
    })]))
  } else {
    fs.writeFileSync(path.join(LINKEDIN_PATH, session.filename), JSON.stringify([Object.assign(profile, {
      picture: '/in/' + path.basename(session.filename).replace('linkedin-', '').replace('.json', '')
        + '/profile.jpg'
    })]))
  }

  res.send('Success!')
}

module.exports = linkedinPic
```
